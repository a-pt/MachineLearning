```
df=data.values.tolist()
Z=np.array(df)
y=Z[:,-1]
X=np.delete(Z, -1, axis=1)
n,d=np.shape(X)
print (n,d)
```

```
Wml=np.matmul(np.matmul(np.linalg.inv(np.matmul(X.T,X)),X.T),y)
```

```
W=np.random.random(100)
result=[]
```

```
for t in range(1,1000):
  step=1e-8
  a1=np.matmul(X.T,X)
  a2=np.matmul(a1,W)
  a3=np.matmul(X.T,y)
  grad=np.subtract(a2,a3)
  W=np.subtract(W,(grad*step))
  diff=0
  for i in range(d):
    diff+=pow((W[i]-Wml[i]),2)
  result.append(np.sqrt(diff))
```

```
plt.title('Distance from Wml Gradient Descent')
plt.xlabel('Iterations')
plt.ylabel('||Wt-Wml||2')
plt.plot(result)
```
