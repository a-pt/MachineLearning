```
def test_train_split(dataset, split=0.70):
	train = list()
	test_size = split * len(dataset)
	test = list(dataset)
	while len(train) < test_size:
		index = randrange(len(test))
		train.append(test.pop(index))
	return train, test
```

```
df=data.values.tolist()
train,test=test_train_split(df,0.7)
print (np.shape(df))
print (np.shape(train))
print (np.shape(test))
```

```
#Preprocessing and Creating Dictionary of Words
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
all_words=[]
for mail in test:
  msg=mail[0].lower()
  for word in re.split('[^a-zA-Z]',msg):
    if(word==''):
      continue
    else:
      all_words.append(word)
print(np.shape(all_words))
all_words = list(dict.fromkeys(all_words))
print(np.shape(all_words))
nltk.download('stopwords')
sw = stopwords.words('english')
all_words = [word for word in all_words if word not in sw]
#Try adding stemmer
stemmer = PorterStemmer()
all_words = [stemmer.stem(word) for word in all_words]
all_words.remove('subject')
all_words = list(dict.fromkeys(all_words))
print(all_words)
Dictionary_spam= dict.fromkeys(all_words,0)
Dictionary_ham= dict.fromkeys(all_words,0)
print(len(all_words))
#print(sorted(Dictionary_spam.items(), key = lambda kv:(kv[1], kv[0])))  
```

```
p=float(spam_count+1)/float(len(train)+2)
print (p)
```

```
p_spam=[]
p_ham=[]
for word in all_words:
  p_spam.append(float(Dictionary_spam[word]+1)/float(spam_count+1))
  p_ham.append(float(Dictionary_ham[word]+1)/float(ham_count+1))
print(p_ham)
print(p_spam)
```

```
#Prediction on Testset
False_positive=0
True_positive=0
False_negative=0
True_negative=0
predict_class=[]
for mail in test:
  frequency =dict.fromkeys(all_words,0)
  mail_words=[]
  for word in re.split('[^a-zA-Z]',mail[0].lower()):
    if(word==''):
      continue
    else:
      mail_words.append(word)
  stemmer = PorterStemmer()
  mail_words = [stemmer.stem(word) for word in mail_words]
  for c_word in all_words:
    if c_word in mail_words:
       frequency[c_word]=1
  f=list(frequency.values()) 
  prob_y_spam=p
  for i in range(0,len(all_words)):
    if(pow((p_spam[i]),f[i]) !=0 and pow((1-p_spam[i]),(1-f[i]))!=0 ):
     prob_y_spam=prob_y_spam*pow((p_spam[i]),f[i])*pow((1-p_spam[i]),(1-f[i]))
  prob_y_ham=(1-p)
  for i in range(0,len(all_words)):
    if(pow((p_ham[i]),f[i]) !=0 and pow((1-p_ham[i]),(1-f[i]))!=0 ):
     prob_y_ham=prob_y_ham*pow((p_ham[i]),f[i])*pow((1-p_ham[i]),(1-f[i]))
  if (prob_y_ham>prob_y_spam):
    predict_class.append(0)
    class_p=0
  else:
    predict_class.append(1)
    class_p=1
  if (mail[1]==0 and class_p==0):
   True_negative+=1
  elif (mail[1]==0 and class_p==1):
   False_positive+=1
  elif (mail[1]==1 and class_p==0):
   False_negative+=1
  elif (mail[1]==1 and class_p==1):
   True_positive+=1
print(True_positive,True_negative,False_positive,False_negative)
```

```
#Performance of Spam Classifier
Precision = True_positive / (True_positive + False_positive)
Recall = True_positive / (True_positive + False_negative)
Fscore = 2 * Precision * Recall / (Precision + Recall)
Accuracy = (True_positive + True_negative) / (True_positive + True_negative + False_positive + False_negative)
print ("Precision :",Precision)
print ("Recall :",Recall)
print ("Fscore :",Fscore)
print ("Accuracy :",Accuracy)
```
