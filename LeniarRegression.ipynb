```
df=data.values.tolist()
Z=np.array(df)
y=Z[:,-1]
X=np.delete(Z, -1, axis=1)
n,d=np.shape(X)
print (n,d)
```

```
W=np.matmul(np.matmul(np.linalg.inv(np.matmul(X.T,X)),X.T),y)
W
```

```
data_test=pd.read_csv('A2Q2Data_test.csv',header=None)
data_test
```

```
df_test=data_test.values.tolist()
Zt=np.array(df_test)
yt=Zt[:,-1]
Xt=np.delete(Zt, -1, axis=1)
n_t,d=np.shape(Xt)
print (n_t,d)
```

```
ypred=np.matmul(Xt,W)
```

```
test_error=0
for i in range(n_t):
  test_error+=(pow((yt[i]-ypred[i]),2))
MSE=test_error/n_t
MSE
```
